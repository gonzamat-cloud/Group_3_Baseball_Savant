```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import os
```

```python
FILE_PATH = "data tools class final data set.csv"  #to read file
df = pd.read_csv(FILE_PATH)

#Looks at file, a peek
df.head()
```

Here is the real good stuff now

First up is EV vs SLG

```python
# Prepare x and y data
x = df["exit_velocity_avg"]
y = df["slg_percent"]

# Clean data 
mask = x.notna() & y.notna()
x = x[mask].values.reshape(-1, 1)
y = y[mask].values

# Create and fit regression model
model = LinearRegression()
model.fit(x, y)
r2 = model.score(x, y)

# Create regression line
x_line = np.linspace(x.min(), x.max(), 200).reshape(-1, 1)
y_line = model.predict(x_line)

# Plot
plt.figure(figsize=(7,5))
plt.scatter(x, y, alpha=0.6)
plt.plot(x_line, y_line, color="red", linewidth=2)
plt.xlabel("Exit Velocity (avg)")
plt.ylabel("Slugging Percentage (SLG)")
plt.title("Exit Velocity vs SLG")
plt.text(0.02, 0.95, f"RÂ² = {r2:.3f}", transform=plt.gca().transAxes, va="top")
plt.tight_layout()
plt.show()
```

Next up is the matrix box for correlations; setup is below

```python
cols_for_corr = [
    "exit_velocity_avg",
    "slg_percent",
    "batting_avg",
    "on_base_percent",
    "b_rbi",
    "isolated_power",
    "on_base_plus_slg"
]

# Converts to numeric in case of text values and compute correlations
corr_df = df[cols_for_corr].apply(pd.to_numeric, errors="coerce")
corr_matrix = corr_df.corr(method="pearson")

# Rounds matrix
corr_matrix.round(3)
```

Next up is the visualization for it

```python
data = corr_matrix.values
labels = corr_matrix.columns

plt.figure(figsize=(7,6))
im = plt.imshow(data, aspect="auto")  # use default color map
plt.colorbar(im, fraction=0.046, pad=0.04)

# Axis labels
plt.xticks(ticks=np.arange(len(labels)), labels=labels, rotation=45, ha="right")
plt.yticks(ticks=np.arange(len(labels)), labels=labels)

# Annotate correlation values in each cell
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        val = data[i, j]
        if not np.isnan(val):
            plt.text(j, i, f"{val:.2f}", ha="center", va="center", color="black")

plt.title("Correlation Matrix for Exit Velocity and Hitting Stats")
plt.tight_layout()
plt.show()
```
